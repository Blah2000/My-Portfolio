```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
url = 'https://www.imdb.com/chart/top/?ref_=nv_mv_250'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.102 Safari/537.36'}
response = requests.get(url, headers=headers)
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
```

```python
response
```

```python
movie_name = []
year = []
time = []
rating = []
vote = []
movie_data = soup.findAll('div', attrs={'class': 'ipc-metadata-list-summary-item__c'})
```

```python
movie_data
```

```python
for i in range(0,9):
    movie_name[i] = movie_name[i][3:]
for i in range(9,99):
    movie_name[i] = movie_name[i][4:]
for i in range(99,250):
    movie_name[i] = movie_name[i][5:]
```

```python
for i in range(0,250):
    print(f"{movie_name[i]} {year[i]} {time[i]} {rating[i]} {vote[i]}")
```

```python
data = {
    'name' : movie_name,
    'year' : year,
    'rating' :rating,
    'rate' : vote,
}
df=pd.DataFrame(data)
df.index = range(1, len(df) + 1)
df.index.name = 'ranking'
df.to_csv('films250.csv', encoding = 'uft-16')
```
